```python
# import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

print(plt.style.available)
```

```python
# read in the data
df = pd.read_csv(r'C:\Users\Zhongxiao Li\Documents\Practice dataset\movies.csv\movies.csv')
```

```python
df.head()
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull()) * 100
    print(f'{col} - {pct_missing:.2f}%')
```

```python
# data types for our columns
df.dtypes
```

```python
# change data type of columns
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df
```

```python
# extract year from released column
def extract_year(x):
    try:
        return str(int(x.split(',')[1].strip().split()[0]))
    except:
        return None
df['yearcorrect'] = df['released'].apply(extract_year)
```

```python
df['yearcorrect'] 
```

```python
# Filling the missing released year from year column
df['yearcorrect'] = df['yearcorrect'].fillna(df['year'])

# Check the missing after the substitution
df['yearcorrect'].isna().sum()
```

```python
df['yearcorrect'] 
```

```python
# Order by gross 
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# check and drop the duplicates
df['company'].drop_duplicates().sort_values(ascending=False)
df.drop_duplicates()
```

```python
# check the correlation with gross
# Scatter plot with budget vs. gross
plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs. Gross Earnings')

plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()
```

```python
# regression plot with budget vs. gross
sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "blue"}, line_kws={"color": "red"})
plt.show()
```

```python
# let's start looking at correlation
df.corr(numeric_only=True) #pearson, kendall, spearman
```

```python
# high correlation between budget and gross
correlation_matrix =  df.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Looks at Compnay
df_numerized = df
for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized 
```

```python
correlation_matrix =  df_numerized.corr(numeric_only=True)
sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Numeric Features')

plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```
